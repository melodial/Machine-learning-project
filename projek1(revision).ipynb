# mendeteksi outlier

```python
#deteksi outlier

def deteksi_outlier(data, column):
    q1 = data[column].quantile(0.25) #kuartil 1
    q3 = data[column].quantile(0.75) #kuartil 3
    IQR = q3 - q1 #rentang antar kuartil
    lower_bound = q1 - 1.5 * IQR #batas bawah
    upper_bound = q3 + 1.5 * IQR #batas atas
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers,lower_bound,upper_bound


kolom_numerik = data.select_dtypes(include=['int64', 'float64']).columns
for i in kolom_numerik:
    outliers,lower_bound,upper_bound = deteksi_outlier(data, i)
    print(f'Kolom {i} memiliki {len(outliers)} outlier dengan batas bawah {lower_bound} dan batas atas {upper_bound}')
    print(f"Batas bawah: {lower_bound}, Batas atas: {upper_bound}\n")
```

menangani outlier dengan robust scaling

```python
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split

#menggunakan robustScaler

scaler = RobustScaler()
data[kolom_numerik] = scaler.fit_transform(data[kolom_numerik])
data.head()
```

```python
for column in kolom_numerik:
    outliers, lower, upper = deteksi_outlier(data, column)
    print(f"Fitur: {column}, Jumlah Outlier: {len(outliers)}")
    print(f"Batas bawah: {lower}, Batas atas: {upper}\n")
```

# uji distribusi menggunakan shapiro-wilk

```python
from scipy import stats

#fungsi untuk melakukan uji shapiro-wilk pada semua kolom
def shapiro_wilk_test(data):
  result = {}
  for column in data.columns:
    statistic, p_value = stats.shapiro(data[column])
    result[column] = {'statistic': statistic, 'p_value': p_value}
  return result

#menjalankan uji dan menampilkan hasilnya
shapiro_results = shapiro_wilk_test(data)

#menampilkan hasil uji untuk setiap kolom
for column, result in shapiro_results.items():
    print(f"Uji Shapiro-Wilk untuk {column}: Statistik = {result['statistic']}, P-value = {result['p_value']}")
    if result['p_value'] > 0.05:
        print(f"{column}: Data terdistribusi normal (gagal menolak hipotesis nol)\n")
    else:
        print(f"{column}: Data tidak terdistribusi normal (menolak hipotesis nol)\n")
```

spearman correlation

```python
from scipy.stats import spearmanr

#menghitung korelasi spearman
spearman = {}
for kolom in X.columns:
    # Mengambil korelasi Spearman antara fitur dan target
    correlation, _ = spearmanr(X[kolom], y)  # Hanya mengambil korelasi, bukan p-value
    spearman[kolom] = correlation

  # Menyortir fitur berdasarkan nilai korelasi dengan abs()
fitur_tersortir = sorted(spearman.items(), key=lambda x: abs(x[1]), reverse=True)

# Menampilkan fitur dan nilai korelasi
for feature, corr in fitur_tersortir:
    print(f"{feature}: {corr}")
```

# implementasi model 1: multiclass logistic reggresion

```python
from sklearn.linear_model import LogisticRegression

#membentuk model regresi logistik untuk muticlass
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')

#melakukan training model
model.fit(X_train_resampled, y_train_resampled)

print(model.score(X_train_resampled, y_train_resampled))
print(model.coef_)
```

```python
#melakukan prediksi terhadap label
predict = model.predict(X_test)
```

```python
#evaluasi model
from sklearn.metrics import classification_report

report = classification_report(y_test, predict)
print(report)
```
