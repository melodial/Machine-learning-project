melakukan pengujian pada distribusi data menggunakan shapiro-wilk

```python
from scipy import stats
#fungsi untuk melakukan uji shapiro-wilk pada semua kolom
def shapiro_wilk_test(data):
    results = {}
    for column in data.columns:
        stat, p_value = stats.shapiro(data[column].dropna())  # dropna() untuk menghindari null values
        results[column] = {'statistic': stat, 'p_value': p_value}
    return results

#menjalankan uji dan menampilkan hasilnya
shapiro_results = shapiro_wilk_test(data)
#menampilkan hasil uji untuk setiap kolom
for column, result in shapiro_results.items():
    print(f"Uji Shapiro-Wilk untuk {column}: Statistik = {result['statistic']}, P-value = {result['p_value']}")
    if result['p_value'] > 0.05:
        print(f"{column}: Data terdistribusi normal (gagal menolak hipotesis nol)\n")
    else:
        print(f"{column}: Data tidak terdistribusi normal (menolak hipotesis nol)\n")
```

feature selection: filter method

```python
from scipy.stats import spearmanr

# Memisahkan fitur X dan target y
X = data.drop(columns=['sub-type for AMI'])
y = data['sub-type for AMI']

# Menghitung korelasi Spearman
spearman = {}
for kolom in X.columns:
    # Mengambil korelasi Spearman antara fitur dan target
    correlation, _ = spearmanr(X[kolom], y)  # Hanya mengambil korelasi, bukan p-value
    spearman[kolom] = correlation

# Menyortir fitur berdasarkan nilai korelasi dengan abs()
fitur_tersortir = sorted(spearman.items(), key=lambda x: abs(x[1]), reverse=True)

# Menampilkan fitur dan nilai korelasi
for feature, corr in fitur_tersortir:
    print(f"{feature}: {corr}")
```

```python
#menghapus fitur dengan korelasi rendah
fitur_terpilih = {k: v for k, v in spearman.items() if abs(v) >= 0.3}
print("fitur terpilih: ",fitur_terpilih)
```

# Implementasi model 1: Logistic regression


```python
# multiclass logistic regression 
from sklearn.linear_model import LogisticRegression

#membentuk model regresi logistik untuk multiclass
model = LogisticRegression(multi_class='multinomial',solver='lbfgs')

model.fit(X_train,y_train)

print(model.coef_)
```

```python
#melakukan prediksi terhadap label
predict = model.predict(X_test)
print(predict)

#untuk  mendapatkan probabilitas prediksi untuk setiap kelas
probabilitas = model.predict_proba(X_test)
print(probabilitas)
```

```python
from sklearn.model_selection import cross_val_score

skor = cross_val_score(model, X,Y, cv=5,scoring='accuracy')
print("cross-validation scores: ",skor)
print("mean skor: ",skor.mean())
```

```python
#evaluasi model
from sklearn.metrics import classification_report

report = classification_report(y_test,predict)
print(report)
```
